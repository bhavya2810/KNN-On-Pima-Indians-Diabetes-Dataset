Question 3
Data Cleaning and Transformation

```python
# Replacing zeros with the NaN, so that we can handle later
input_data[['Glucose', 'BloodPressure', 'SkinThickness','Insulin','BMI']]=input_data[['Glucose', 'BloodPressure', 'SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
input_data.isnull().sum() # After replacing with NaN, we can see that there are null values in the dataset
```

```python
input_data.describe()
```

```python
input_data.fillna(input_data.mean(), inplace = True) #To fill null values with mean values of the data
```

```python
input_data.isnull().sum() # To verify if there are any null values after handling
```

Handling Outliers

```python
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(input_data))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

```python
input_data_after_outliers_removal = input_data[(z < 3).all(axis=1)]
input_data_after_outliers_removal.to_csv('data/samples/processed_data.csv')
```

Question 4
KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
x = input_data_after_outliers_removal[['Pregnancies', 'Glucose',
       'BloodPressure', 'SkinThickness',
       'Insulin', 'BMI', 'DiabetesPedigreeFunction',
       'Age']]
y = input_data_after_outliers_removal['Outcome']
x_train_data, x_test_data, y_train_data, y_test_data = train_test_split(x, y, random_state=1, test_size=0.3333)
```

```python
x_train_data.head(2)
```

```python
y_train_data.head(2)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_model.fit(x_train_data, y_train_data)
y_predicted = knn_model.predict(x_test_data)
y_predicted
```

Cross Validation

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cross_validation_scores = []
neighbors = list(np.arange(2,50,1))
for value in neighbors:
    knn_model = KNeighborsClassifier(n_neighbors = value, metric='euclidean')
    cross_validatoion = cross_val_score(knn_model, x_train_data, y_train_data, cv = 5 , scoring = 'accuracy')
    cross_validation_scores.append(cross_validatoion.mean())
error = [1-x for x in cross_validation_scores]
optimal_k = neighbors[ error.index(min(error)) ]
knn_optimal = KNeighborsClassifier(n_neighbors = optimal_k, metric='euclidean')
knn_optimal.fit(x_train_data, y_train_data)
y_prediction = knn_optimal.predict(x_test_data)
accuracy = accuracy_score(y_test_data, y_prediction)*100
print("The accuracy for optimal k = {0} using euclidean is {1}".format(optimal_k, accuracy))
```

Loss Functions
Mean Square Error

```python
MSE = np.square(np.subtract(y_test_data,y_predicted)).mean() 
MSE
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test_data, y_predicted) 
```

Cross Entropy loss

```python
def cross_entropy(predictions, targets, epsilon=1e-10):
    predictions = np.clip(predictions, epsilon, 1. - epsilon)
    N = predictions.shape[0]
    ce_loss = -np.sum(np.sum(targets * np.log(predictions + 1e-5)))/N
    return ce_loss
cross_entropy_loss = cross_entropy(y_predicted, y_test_data)
print ("Cross entropy loss is: " + str(cross_entropy_loss))
```
